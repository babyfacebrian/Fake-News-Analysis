```python
def run_ner_parse(corpus: List[str]):
    nlp = spacy.load('en_core_web_md')
    ner_pipeline = nlp.pipe(corpus, disable=["tagger", "parser"])
    ner_types = ['PERSON', 'NORP', 'FAC', 'ORG', 'GPE', 'PRODUCT']

    ner_corpus = []
    for doc in ner_pipeline:
        found_entities = []
        for entity in doc.ents:
            if entity.label_ in ner_types:
                found_entities.append(entity.text)
        if len(found_entities) > 0:
            ner_corpus.append(' '.join(found_entities))
    return ner_corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

def compute_doc_frequency_matrix(data):
    vectorizer = CountVectorizer()
    docs = vectorizer.fit_transform(data)
    words = vectorizer.get_feature_names()
    
    matrix = []
    for arr in docs.toarray():
        row = dict(zip(words, arr))
        matrix.append(row)
    return pd.DataFrame(matrix)


def compute_inverse_doc_matrix(data):
    vectorizer = TfidfVectorizer()
    docs = vectorizer.fit_transform(data)
    words = vectorizer.get_feature_names()
    
    matrix = []
    for arr in docs.toarray():
        row = dict(zip(words, arr))
        matrix.append(row)
    return pd.DataFrame(matrix)
    
```
